# Netflix Movie Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mymoviedb.csv', lineterminator='\n')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
df
```

```python
df['Release_Date'].dtype
```

```python
df.drop(columns=['Overview', 'Original_Language', 'Poster_Url'], inplace=True)
```

```python
df.columns
```

```python
df['Vote_Average'].value_counts()
```

```python
df['Genre'] = df['Genre'].str.split(', ')
```

```python
df = df.explode('Genre').reset_index(drop=True)
```

```python
df
```

```python
df['Genre'] = df['Genre'].astype('category')
```

```python
df.nunique()
```

```python
sns.set_style('whitegrid')
```

### Most frequent genre of movies released on Netflix

```python
df['Genre'].describe()
```

```python
df['Genre'].value_counts().plot(kind='barh',title='Genre DistributionTable', ylabel="Gener", xlabel='Count',grid=False)
plt.show()
```

```python
df['Vote_Average'].describe()
```

```python
df.sort_values(by='Vote_Count',ascending=False)
```

### Highest votes in the vote_avg column.

```python
df['VAG'] = df['Vote_Average'].apply(lambda x: 'Popular' if x > 7.5 else ('Average' if x > 5.0 else ('Below_avg' if x > 2.5 else 'Not_popular')))
```

```python
df['VAG'].value_counts().plot(kind='pie',title='Highest votes category', ylabel="lables", xlabel='Count',grid=False)
plt.show()
```

### Movie with highest popularity

```python
df.sort_values(by='Popularity', ascending=False)
```

```python
df[df['Popularity'] == df['Popularity'].max()]
```

### Movie with lowest popularity

```python
df[df['Popularity'] == df['Popularity'].min()]
```
